```python
# 📌 라이브러리 import & 폰트 설정
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='NanumGothic')  # 한글 폰트 설정 (Colab에서는 일부 작동 안 할 수 있음)

# 📌 샘플 데이터 (아이폰, 갤럭시탭, 에어팟)
data = {
    "product_name": [
        "아이폰 13 Pro", "갤럭시탭 S7", "에어팟 프로", "아이폰 12", "갤럭시탭 A8",
        "에어팟 2세대", "아이폰 11", "아이폰 SE2", "갤럭시탭 S6", "에어팟 맥스"
    ],
    "price": [850000, 400000, 160000, 600000, 300000, 100000, 450000, 350000, 380000, 650000],
    "date": [
        "2025-04-22", "2025-04-21", "2025-04-21", "2025-04-20", "2025-04-19",
        "2025-04-19", "2025-04-18", "2025-04-18", "2025-04-17", "2025-04-17"
    ],
    "description": [
        "박스포함 상태 좋음", "태블릿 단독 판매", "정품 보증 남음", "생활기스 있음", "깨끗하게 사용",
        "한쪽 소리 작음", "배터리 90%", "유심 호환 문제 없음", "가죽 케이스 포함", "미개봉 새제품"
    ]
}

df = pd.DataFrame(data)
df.head()
# Calculate average price
avg_price = df['price'].mean()
print(f"📌 Average Price: ₩{avg_price:,.0f}\n")

# Above average
print("📈 Products Above Average Price:")
for idx, row in df[df['price'] > avg_price].iterrows():
    print(f" - {row['product_name']}: ₩{row['price']:,}") # Access 'price' from the row

# Below average
print("\n📉 Products Below Average Price:")
for idx, row in df[df['price'] < avg_price].iterrows():
    print(f" - {row['product_name']}: ₩{row['price']:,}") # Access 'price' from the row
```

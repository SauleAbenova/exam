```python
# üìå ÎùºÏù¥Î∏åÎü¨Î¶¨ import & Ìè∞Ìä∏ ÏÑ§Ï†ï
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rc('font', family='NanumGothic')  # ÌïúÍ∏Ä Ìè∞Ìä∏ ÏÑ§Ï†ï (ColabÏóêÏÑúÎäî ÏùºÎ∂Ä ÏûëÎèô Ïïà Ìï† Ïàò ÏûàÏùå)

# üìå ÏÉòÌîå Îç∞Ïù¥ÌÑ∞ (ÏïÑÏù¥Ìè∞, Í∞§Îü≠ÏãúÌÉ≠, ÏóêÏñ¥Ìåü)
data = {
    "product_name": [
        "ÏïÑÏù¥Ìè∞ 13 Pro", "Í∞§Îü≠ÏãúÌÉ≠ S7", "ÏóêÏñ¥Ìåü ÌîÑÎ°ú", "ÏïÑÏù¥Ìè∞ 12", "Í∞§Îü≠ÏãúÌÉ≠ A8",
        "ÏóêÏñ¥Ìåü 2ÏÑ∏ÎåÄ", "ÏïÑÏù¥Ìè∞ 11", "ÏïÑÏù¥Ìè∞ SE2", "Í∞§Îü≠ÏãúÌÉ≠ S6", "ÏóêÏñ¥Ìåü Îß•Ïä§"
    ],
    "price": [850000, 400000, 160000, 600000, 300000, 100000, 450000, 350000, 380000, 650000],
    "date": [
        "2025-04-22", "2025-04-21", "2025-04-21", "2025-04-20", "2025-04-19",
        "2025-04-19", "2025-04-18", "2025-04-18", "2025-04-17", "2025-04-17"
    ],
    "description": [
        "Î∞ïÏä§Ìè¨Ìï® ÏÉÅÌÉú Ï¢ãÏùå", "ÌÉúÎ∏îÎ¶ø Îã®ÎèÖ ÌåêÎß§", "Ï†ïÌíà Î≥¥Ï¶ù ÎÇ®Ïùå", "ÏÉùÌôúÍ∏∞Ïä§ ÏûàÏùå", "Íπ®ÎÅóÌïòÍ≤å ÏÇ¨Ïö©",
        "ÌïúÏ™Ω ÏÜåÎ¶¨ ÏûëÏùå", "Î∞∞ÌÑ∞Î¶¨ 90%", "Ïú†Ïã¨ Ìò∏Ìôò Î¨∏Ï†ú ÏóÜÏùå", "Í∞ÄÏ£Ω ÏºÄÏù¥Ïä§ Ìè¨Ìï®", "ÎØ∏Í∞úÎ¥â ÏÉàÏ†úÌíà"
    ]
}

df = pd.DataFrame(data)
df.head()
# Calculate average price
avg_price = df['price'].mean()
print(f"üìå Average Price: ‚Ç©{avg_price:,.0f}\n")

# Above average
print("üìà Products Above Average Price:")
for idx, row in df[df['price'] > avg_price].iterrows():
    print(f" - {row['product_name']}: ‚Ç©{row['price']:,}") # Access 'price' from the row

# Below average
print("\nüìâ Products Below Average Price:")
for idx, row in df[df['price'] < avg_price].iterrows():
    print(f" - {row['product_name']}: ‚Ç©{row['price']:,}") # Access 'price' from the row
```

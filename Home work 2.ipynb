```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Misol uchun, DataFrame
data = {
    'product_name': ['iPhone 12', 'iPhone 12', 'Galaxy Tab', 'AirPods', 'iPhone 13'],
    'price': [600, 650, 500, 150, 700],
    'date': ['2025-04-10', '2025-04-15', '2025-04-14', '2025-04-16', '2025-04-18'],
    'description': ['Good condition', 'New', 'Slightly used', 'Brand new', 'Excellent condition']
}
df = pd.DataFrame(data)

# Histogram
plt.figure(figsize=(10,6))
sns.histplot(df['price'], kde=True, bins=5)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

# Boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x=df['price'])
plt.title('Price Boxplot')
plt.show()
```

```python
avg_price = df['price'].mean()
print(f"Average Price: {avg_price}")

above_avg = df[df['price'] > avg_price]
below_avg = df[df['price'] < avg_price]

print("Above Average Price Products:")
print(above_avg)

print("Below Average Price Products:")
print(below_avg)
```

```python
sns.pairplot(df[['price', 'date', 'description']])
plt.show()
```

```python
df.to_csv('used_product_data.csv', index=False)
```

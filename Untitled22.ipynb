```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
keyword = "인공지능"
url = f"https://search.naver.com/search.naver?where=news&query={keyword}"
headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

```

```python
titles = []
dates = []
links = []

news_list = soup.select(".news_tit")

for item in news_list:
    title = item.get("title")
    link = item.get("href")

    # Даты у Naver часто не отображаются — можно оставить пусто или парсить из другого места
    date = "알 수 없음"  # "Неизвестно"

    titles.append(title)
    links.append(link)
    dates.append(date)
```

```python
df = pd.DataFrame({
    "제목": titles,
    "날짜": dates,
    "링크": links
})

df.to_csv("naver_news.csv", index=False)
```

```python
import feedparser

keyword = "인공지능"
rss_url = f"https://news.google.com/rss/search?q={keyword}+site:naver.com&hl=ko&gl=KR&ceid=KR:ko"

feed = feedparser.parse(rss_url)

titles = []
links = []
dates = []

for entry in feed.entries:
    titles.append(entry.title)
    links.append(entry.link)
    dates.append(entry.published)

import pandas as pd

df = pd.DataFrame({
    "제목": titles,
    "날짜": dates,
    "링크": links
})

df.to_csv("naver_news_rss.csv", index=False)
print("Сохранено новостей:", len(df))
```

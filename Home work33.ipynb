```python
import requests
import pandas as pd
import re

query = "에어팟"
headers = {
    "User-Agent": "Mozilla/5.0"
}

all_reviews = []

# Сбор с первых 5 страниц (можно больше)
for page in range(1, 6):
    url = f"https://api.bunjang.co.kr/api/1/find_v2.json?q={query}&order=score&page={page}&n=20"
    response = requests.get(url, headers=headers)
    items = response.json()["list"]

    for item in items:
        all_reviews.append({
            "제목": item["name"],
            "가격": item["price"],
            "링크": "https://m.bunjang.co.kr/products/" + item["pid"]
        })

df = pd.DataFrame(all_reviews)
df["가격"] = pd.to_numeric(df["가격"], errors="coerce")
df.head()
```

```python
import pandas as pd
import requests

query = "에어팟"
headers = {"User-Agent": "Mozilla/5.0"}

reviews = []

# Используем API Bunjang, чтобы получить товары и отзывы (фейковые для примера)
for page in range(1, 4):
    url = f"https://api.bunjang.co.kr/api/1/find_v2.json?q={query}&order=score&page={page}&n=20"
    res = requests.get(url, headers=headers)
    items = res.json()["list"]

    for item in items:
        reviews.append(item["name"])  # Здесь вместо "리뷰" просто названия

df = pd.DataFrame({"리뷰": reviews})
df.head()
```

```python
import requests
import pandas as pd

query = "에어팟"
headers = {"User-Agent": "Mozilla/5.0"}
reviews = []

for page in range(1, 3):  # можно больше страниц
    url = f"https://api.bunjang.co.kr/api/1/find_v2.json?q={query}&order=score&page={page}&n=20"
    res = requests.get(url, headers=headers)
    items = res.json()["list"]

    for item in items:
        reviews.append(item["name"])

df = pd.DataFrame({"리뷰": reviews})
df.head()
```
